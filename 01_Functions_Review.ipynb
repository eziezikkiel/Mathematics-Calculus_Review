```python
# Import libraries needed
import numpy as np 
import matplotlib.pyplot as plt 
```

I will create my own function to plot everything quickly.

```python
# Create a function that will plot the mathematical functions
def plot_func(x,y,label=None):
    
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.grid()
    if label == None:
        plt.plot(x,y);
    else:
        plt.plot(x,y,label=label)
        plt.legend();
```

```python
# Create x and y values
x = np.linspace(-5, 5, 100)
y = 0.5 * x + 1

# Plot line
plot_func(x,y)
```

```python
# Create x and y values for quadratic
x = np.linspace(-2, 2, 100)
y = -0.5 * 9.8 * x ** 2 + 2 * x + 1

# Plot quadratic
plot_func(x,y);
```

```python
# Create x and y for root funcion
x = np.linspace(0, 10, 100)
y = -0.5 * x ** (1/3) + 0.2 * x ** (1/2) + 1

# Plot root function
plot_func(x,y);
```

```python
# Create x and y for exponential function
x = np.linspace(-5, 5, 100)
y = np.exp(x) # e ** x

# Plot exponential function
plot_func(x,y);
```

```python
# Compare exponential to quadratic

# Create x and y
x = np.linspace(-5, 5, 100)
y = np.exp(x)
y2 = x**2

# Plot both functions
plot_func(x,y,label='e^x')
plot_func(x,y2, label='e^2');
```

```python
# Create x and y for log functions
x = np.linspace(0.0000001, 10, 200)
y = np.log(x) # ln(x)

# Plot log function
plot_func(x,y);
```

```python
# Create x and y for sine function
x = np.linspace(-5 * np.pi, 5 * np.pi, 200)
y = np.sin(x)

# Plot sine
plot_func(x,y)
```

```python
# Create x and y for sine & cosine function
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.sin(x)
y2 = np.cos(x)

# Plot sine and cosine
plot_func(x,y,label='sin')
plot_func(x,y2,label='cos')
```

```python
# Create x and y for tan function
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.tan(x)

# Plot tan function
plot_func(x,y)
```

```python
# Create x and y for tan function
x = np.linspace(-2 * np.pi, 2 * np.pi, 100_000) # with 100,000 points
y = np.tan(x)

# Plot tan function
plot_func(x,y)
```

```python
# Create x and y for hyperbolic sine
x = np.linspace(-5, 5, 100)
y = np.sinh(x)

# Plot hyperbolic sine
plot_func(x,y)
```

```python
# Create x and y for hyperbolic sine and cosine
x = np.linspace(-5, 5, 100)
y = np.sinh(x)
y2 = np.cosh(x)

# Plot hyperbolic sine and cosine
plot_func(x,y,label='sinh')
plot_func(x,y2,label='cosh')
```

**Question:** Does sinh and cosh equal on first quadrant or close?

To answer this we will perform equalities in python.

```python
# Compare y to y2
y == y2
```

We can see that on all points it's false meaning they're not equal, let's check how close they are to each other.

To do that we will subtract each other from the 50th point using list slicing.

```python
y[50:] - y2[50:]
```

> We can conclude that although they look like they overlap, in reality they're just very close.

```python
# Create x and y for hyperbolic tangent
x = np.linspace(-5, 5, 100)
y = np.tanh(x)

# Plot hyperbolic tangent
plot_func(x,y)
```
